```python
def load_training_summaries(name):    
    basepath = '/mnt/ceph/users/fmaion/projects/bias_priors/training_results/'+name+'/'
    
    with open(basepath+name+'_posterior.pkl', 'rb') as f:
        # This loads the object exactly as it was saved
        posterior_ensemble = pickle.load(f)
    
    with open(basepath+name+'_summary.json', 'rb') as f:
        summaries = json.load(f)

    return posterior_ensemble, summaries
```

```python
posterior, summaries = load_training_summaries('bias_NPE')
posterior_int, summaries_int = load_training_summaries('bias_NPE_intermediate_GEM')
posterior_std, summaries_std = load_training_summaries('bias_NPE_large_std')
posterior_bs_v2, summaries_bs_v2 = load_training_summaries('bias_NPE_large_bs_v2')
posterior_std_bs_v2, summaries_std_bs_v2 = load_training_summaries('bias_NPE_large_std_bs_v2')
posterior_tiny_std_bs_v2, summaries_tiny_std_bs_v2 = load_training_summaries('bias_NPE_tiny_std_bs_v2')
posterior_std_bs_v2_wd, summaries_std_bs_v2_wd = load_training_summaries('bias_NPE_large_std_bs_v2_wd')
posterior_std_11_bs_v2, summaries_std_11_bs_v2 = load_training_summaries('bias_NPE_large_std_11_bs_v2')
```

```python
# plot train/validation loss
fig, ax = plt.subplots(1, 1, figsize=(5.5, 5), dpi=200)

# Large Model Standard
ax.plot(np.array(summaries_std[0]['training_log_probs']),
        ls='-', label="Large Model Std", color="C0")
ax.plot(np.array(summaries_std[0]['validation_log_probs']),
        ls=':', color="C0")

# Tiny Model Standard
ax.plot(np.array(summaries_tiny_std_bs_v2[0]['training_log_probs']),
        ls='-', label="Tiny Model Std", color="C6")
ax.plot(np.array(summaries_tiny_std_bs_v2[0]['validation_log_probs']),
        ls=':', color="C6")

# Large Model Standardized + small batch-size
ax.plot(np.array(summaries_std_bs_v2[0]['training_log_probs']),
        ls='-', label="Large Model Std with batch-size=256", color="C2")
ax.plot(np.array(summaries_std_bs_v2[0]['validation_log_probs']),
        ls=':', color="C2")

# # Large Model Standardized + small batch-size + weight decay
# ax.plot(np.array(summaries_std_bs_v2_wd[0]['training_log_probs']),
#         ls='-', label="Large Model Std with batch-size=256", color="C3")
# ax.plot(np.array(summaries_std_bs_v2_wd[0]['validation_log_probs']),
#         ls=':', color="C3")

# Larg Model Reduced Parameters
ax.plot(np.array(summaries_std_11_bs_v2[0]['training_log_probs']),
        ls='-', label="Large Model with 11 parameters", color="C3")
ax.plot(np.array(summaries_std_11_bs_v2[0]['validation_log_probs']),
        ls=':', color="C3")

# Large Model
ax.plot(np.array(summaries[0]['training_log_probs']),
        ls='-', label="Large Model", color="k")
ax.plot(np.array(summaries[0]['validation_log_probs']),
        ls=':', color="k")

# Large model + smaller batch-size
ax.plot(np.array(summaries_bs_v2[0]['training_log_probs']),
        ls='-', label="Large Model with batch-size=256", color="C1")
ax.plot(np.array(summaries_bs_v2[0]['validation_log_probs']),
        ls=':', color="C1")

ax.plot(np.zeros(10), np.zeros(10), color='k', label="Training Loss")
ax.plot(np.zeros(10), np.zeros(10), color='k', ls=':', label="Validation Loss")
ax.set_ylim(-1.6,-0.5)

ax.set_xlim(0)
ax.set_xlabel('Epoch')
ax.set_ylabel('Log probability')
ax.legend(fontsize=8, loc='lower right')

plt.grid()
```
