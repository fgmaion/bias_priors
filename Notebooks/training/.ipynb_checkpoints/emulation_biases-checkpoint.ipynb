```python
def binned_percentile_errorbar(
    x, y, nbins=10, percentiles=(16, 50, 84), ax=None, **errorbar_kwargs
):
    """
    Bin points by x, compute specified percentiles of y in each bin, and plot as asymmetric errorbars.

    Parameters
    - x, y: array-like, same length
    - nbins: int, number of x-bins
    - percentiles: sequence of percentiles (e.g. [16,50,84]). Must have at least 2 values.
      The central value used for the point is the middle percentile after sorting;
      the errorbars use the min and max percentiles (first and last after sorting).
    - ax: matplotlib Axes (defaults to current axes)
    - show_empty: if True, will plot bins with no points as NaN (not shown); otherwise they are skipped
    - errorbar_kwargs: additional kwargs forwarded to ax.errorbar

    Returns
    - dict with keys:
      'bin_centers', 'center_vals', 'err_low', 'err_high', 'bin_edges', 'counts'
    """
    x = np.asarray(x)
    y = np.asarray(y)
    if x.shape != y.shape:
        raise ValueError("x and y must have the same shape")

    # filter NaNs
    mask = ~np.isnan(x) & ~np.isnan(y)
    x = x[mask]
    y = y[mask]

    if x.size == 0:
        raise ValueError("No valid (non-NaN) points to bin")

    if nbins < 1:
        raise ValueError("nbins must be >= 1")

    p = np.sort(np.asarray(percentiles))
    if p.size < 2:
        raise ValueError("percentiles must contain at least two values")

    ax = ax if ax is not None else plt.gca()

    bin_edges = np.linspace(x.min(), x.max(), nbins + 1)
    bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
    inds = np.digitize(x, bin_edges) - 1  # bins numbered 0..nbins-1

    center_vals = np.full(nbins, np.nan)
    err_low = np.full(nbins, np.nan)
    err_high = np.full(nbins, np.nan)
    counts = np.zeros(nbins, dtype=int)

    for i in range(nbins):
        ybin = y[inds == i]
        counts[i] = ybin.size
        if ybin.size > 0:
            qs = np.percentile(ybin, p)
            mid_idx = len(qs) // 2
            center = qs[mid_idx]
            low = center - qs[0]
            high = qs[-1] - center
            center_vals[i] = center
            err_low[i] = low
            err_high[i] = high

    valid = ~np.isnan(center_vals)
    plot_centers = bin_centers[valid]
    plot_centers_vals = center_vals[valid]
    plot_err_low = err_low[valid]
    plot_err_high = err_high[valid]

    if plot_centers.size > 0:
        yerr = np.vstack([plot_err_low, plot_err_high])
        ax.errorbar(plot_centers, plot_centers_vals,
                    yerr=yerr, **errorbar_kwargs)
```

## Construct and visualize a mock simulator

```python
def simulator(omega_m, sigma8, logmstar):
    # A mock simulator that returns bias as a function of cosmology and Mstar
    mstar = 10**logmstar
    b1 = 1 + 0.5 * (mstar / 1e11)**0.5 * \
        (sigma8 / 0.8)**1 * (omega_m / 0.3)**0.6
    b2 = 1 + 0.1 * (mstar / 1e11)**0.3 * \
        (sigma8 / 0.8)**1 * (omega_m / 0.3)**0.2
    # Add some noise
    n1 = np.random.normal(loc=0.0, scale=0.05, size=np.shape(b1)) * b1
    b1 = b1 + n1
    n2 = np.random.normal(loc=0.0, scale=0.05, size=np.shape(b2)) * b2
    b2 = b2 + n2
    return np.array([b1, b2])
```

```python
prior_range = [
    (0.1, 0.5),  # omega_m
    (0.6, 1.0),  # sigma8
    (8.5, 12.0)  # logmstar
]
low, high = np.array(prior_range).T
```

```python
# two-panel plot: left varies omega_m, right varies sigma8
om_vals = np.linspace(low[0], high[0], 5)
sigma8_vals = np.linspace(low[1], high[1], 5)
sigma8_fixed = 0.8
om_fixed = 0.3  # representative omega_m for the sigma8 panel
logm_grid = np.linspace(*prior_range[-1], 1000)

fig, axs = plt.subplots(2, 2, figsize=(10, 6), sharex=True)

# left: vary omega_m

for om in om_vals:
    b_pred = simulator(om, sigma8_fixed, logm_grid)
    binned_percentile_errorbar(
        logm_grid, b_pred[0], nbins=15, percentiles=(16, 50, 84),
        ax=axs[0, 0], fmt='o-', color=plt.cm.viridis((om - low[0]) / (high[0] - low[0])),
        label=f'Om={om:.2f}',
    )
    binned_percentile_errorbar(
        logm_grid, b_pred[1], nbins=15, percentiles=(16, 50, 84),
        ax=axs[1, 0], fmt='s-', color=plt.cm.viridis((om - low[0]) / (high[0] - low[0])),
        label=f'Om={om:.2f} (b2)',
    )
axs[0, 0].set(ylabel="b_1",
              title=f'Varying Omega_m (sigma8={sigma8_fixed:.2f})')
axs[0, 0].legend(title="Omega_m")
axs[1, 0].set(xlabel="log10(Mstar)", ylabel="b_2")

# right: vary sigma8
for s8 in sigma8_vals:
    b_pred = simulator(om_fixed, s8, logm_grid)
    binned_percentile_errorbar(
        logm_grid, b_pred[0], nbins=15, percentiles=(16, 50, 84),
        ax=axs[0, 1], fmt='o-', color=plt.cm.plasma((s8 - low[1]) / (high[1] - low[1])),
        label=f'S8={s8:.2f}',
    )
    binned_percentile_errorbar(
        logm_grid, b_pred[1], nbins=15, percentiles=(16, 50, 84),
        ax=axs[1, 1], fmt='s-', color=plt.cm.plasma((s8 - low[1]) / (high[1] - low[1])),
        label=f'S8={s8:.2f} (b2)',
    )
axs[0, 1].set(title=f'Varying sigma8 (Om={om_fixed:.2f})')
axs[0, 1].legend(title="sigma8")
axs[1, 1].set(xlabel="log10(Mstar)")

plt.tight_layout()
plt.show()
```

## Make a training dataset

```python
prior_range = [
    (0.1, 0.5),  # omega_m
    (0.6, 1.0),  # sigma8
    (8.5, 12.0)  # logmstar
]
low, high = np.array(prior_range).T

# draw 2000 cosmologies
Nsims = 2000
cosmo = np.random.uniform(low[:2], high[:2], size=(Nsims, low.shape[0]-1))
ids = np.arange(Nsims)

# draw 50 logmstar values per cosmology
Nlogm = 50
logmstar = np.random.uniform(low[2], high[2], (Nsims*Nlogm))

# collate cosmology parameters to match logmstar shape
omega_m = np.repeat(cosmo[:, 0], Nlogm)
sigma8 = np.repeat(cosmo[:, 1], Nlogm)
ids = np.repeat(ids, Nlogm)

# combine all into a theta vector
theta = np.vstack([omega_m, sigma8, logmstar]).T

# call simulator
b1, b2 = simulator(theta[:, 0], theta[:, 1], theta[:, 2])

# combine all into an x vector
x = np.vstack([b1, b2]).T
```

```python
# print shapes
print("theta shape:", theta.shape)
print("x shape:", x.shape)
print("ID shape:", ids.shape)

# TODO: You can replace theta, x, and ids with your own data
```

```python
# let's plot a few examples

f, axs = plt.subplots(1, 2, figsize=(10, 4))


for i in np.random.randint(0, Nsims, 4):
    # grab data for this ID
    t_ = theta[ids == i]
    x_ = x[ids == i]
    b1_, b2_ = x_.T
    logmstar_ = t_[:, 2]

    ax = axs[0]
    binned_percentile_errorbar(
        logmstar_, b1_, nbins=10, percentiles=(16, 50, 84),
        ax=ax, fmt='o-', label=f'ID={i}',
    )

    ax = axs[1]
    binned_percentile_errorbar(
        logmstar_, b2_, nbins=10, percentiles=(16, 50, 84),
        ax=ax, fmt='s-', label=f'ID={i}',
    )
axs[0].set(xlabel="log10(Mstar)", ylabel="b_1")
axs[1].set(xlabel="log10(Mstar)", ylabel="b_2")
axs[0].legend()
```

```python
# create train/test split, splitting on ids
from sklearn.model_selection import train_test_split

test_frac = 0.1
train_ids, test_ids = train_test_split(
    np.unique(ids), test_size=test_frac, random_state=42
)

theta_train = theta[np.isin(ids, train_ids)]
x_train = x[np.isin(ids, train_ids)]
theta_test = theta[np.isin(ids, test_ids)]
x_test = x[np.isin(ids, test_ids)]

print("Train length:", theta_train.shape[0])
print("Test length:", theta_test.shape[0])
```

## Training example model with ILI

```python
# set up a dataloader
# NOTE: these are swapped because we want to do an emulation task
loader = NumpyLoader(x=theta_train, theta=x_train)
```

```python
device = 'cpu'  # or 'cuda'

# define a prior  (infinite for an emulation task)
prior = ili.utils.Uniform(low=[-np.inf, -np.inf],
                          high=[np.inf, np.inf], device=device)

# instantiate your neural networks to be used as an ensemble
nets = [
    ili.utils.load_nde_lampe(
        engine='NPE', model='nsf', hidden_features=8, num_transforms=4)
]

# define training arguments
train_args = {
    'training_batch_size': 64,
    'learning_rate': 1e-4
}

# initialize the trainer
runner = InferenceRunner.load(
    backend='lampe',
    engine='NPE',
    prior=prior,
    nets=nets,
    device=device,
    train_args=train_args,
    proposal=None,
    out_dir=None
)
```

```python
# train the model
posterior_ensemble, summaries = runner(loader=loader)
```

```python
# plot train/validation loss
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
for i, m in enumerate(summaries):
    ax.plot(np.array(m['training_log_probs']),
            ls='-', label=f"{i}_train")
    ax.plot(np.array(m['validation_log_probs']),
            ls='--', label=f"{i}_val")
ax.set_xlim(0)
ax.set_xlabel('Epoch')
ax.set_ylabel('Log probability')
ax.legend()
```

```python
# generate samples for a single, given test point
idx = 9
t_ = theta_test[idx]
x_ = x_test[idx]

print(f'Omega_m:, {t_[0]:.3f}')
print(f'sigma8: {t_[1]:.3f}')
print(f'logmstar: {t_[2]:.3f}')

samples = posterior_ensemble.sample((1000,), t_)

f, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.scatter(samples[:, 0], samples[:, 1], label='Emulator samples',
           marker='.', alpha=0.8)
ax.plot(x_[0], x_[1], 'rx',
        markersize=12, label='True value')
ax.legend()
ax.set(xlabel='b1', ylabel='b2')
```

```python
# now, let's sample a range of mstar values for a fixed cosmology
idx = 0
id_sim = test_ids[idx]
theta_sim = theta_test[ids[np.isin(ids, test_ids)] == id_sim]
x_sim = x_test[ids[np.isin(ids, test_ids)] == id_sim]

Nsamp = 100
samples = [
    posterior_ensemble.sample((Nsamp,), t_, show_progress_bars=False) for t_ in theta_sim]
samples = np.array(samples)
logmstar_ = theta_sim[:, 2].repeat(Nsamp)

f, axs = plt.subplots(2, 1, figsize=(8, 10), sharex=True)

ax = axs[0]
binned_percentile_errorbar(
    logmstar_, samples[:, :, 0].flatten(), nbins=10, percentiles=(16, 50, 84),
    ax=ax, fmt='o-', color='C0',
    label='Emulator samples',
)
binned_percentile_errorbar(
    theta_sim[:, 2], x_sim[:, 0], nbins=10, percentiles=(16, 50, 84),
    ax=ax, fmt='s--', color='C1',
    label='True values',
)
ax.set(ylabel="b_1")
ax.legend()

ax = axs[1]
binned_percentile_errorbar(
    logmstar_, samples[:, :, 1].flatten(), nbins=10, percentiles=(16, 50, 84),
    ax=ax, fmt='o-', color='C0',
    label='Emulator samples',
)
binned_percentile_errorbar(
    theta_sim[:, 2], x_sim[:, 1], nbins=10, percentiles=(16, 50, 84),
    ax=ax, fmt='s--', color='C1',
    label='True values',
)
ax.set(xlabel="log10(Mstar)", ylabel="b_2")
ax.legend()
```
